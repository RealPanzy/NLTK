```python
import nltk
from nltk.tokenize import word_tokenize
import nltk.corpus
from nltk.probability import FreqDist

text = "Want help scanning the TempState folder or converting recovered files? I can walk you through it step by step"
tokenized_text = word_tokenize(text)
print(tokenized_text)
print("Token count:", len(tokenized_text))

fdist=FreqDist()
for i in tokenized_text:
    fdist[i.lower()]+=1
fdist

```

```python
fdist['help']
```

```python
most_occuring=fdist.most_common(100)
most_occuring
```

```python
text1="Under a flickering streetlamp in the forgotten part of town, a broken vending machine hummed quietly beside an alley full of secrets. It never sold soda or chips — only handwritten notes, fortunes left behind by strangers, and once, a paper crane folded from a train ticket to nowhere. Moths danced like thoughts above its rusted roof, and every Thursday at midnight, someone in a green coat left a single daffodil beneath it, without fail."
tk2=word_tokenize(text1)
```

```python
bgram=list(nltk.bigrams(tk2))
bgram
```

```python
ngram=list(nltk.ngrams(tk2,6))
ngram
```

```python
#stemming
from nltk.stem import PorterStemmer
pst=PorterStemmer()
for a in tk2:
    print(pst.stem(a))
```

```python
#lemmatization
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
word_lem=WordNetLemmatizer()
for b in tk2:
    print(b+ " :"+word_lem.lemmatize(b))
```

```python
import re

pattern = r"[a-zA-z]+" 

for gram in ngram:
    gram_str = ' '.join(gram) 
    if re.search(pattern, gram_str):
        print("Match found in:", gram_str)
```
